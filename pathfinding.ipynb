# Pathfinding AI: From Zero to Hero

Welcome to Classical AI. To a computer, a map is not a picture; it is a mathematical structure called a **Graph**. A graph consists of **Nodes** (intersections or grid squares) and **Edges** (the roads or paths connecting them).

### The Mathematical Core: $f(n) = g(n) + h(n)$
The A* algorithm evaluates which node to explore next based on this formula:
* $g(n)$: The exact cost to reach node $n$ from the start.
* $h(n)$: The **heuristic** (estimated cost) from node $n$ to the goal.
* $f(n)$: The total estimated cost of the path through node $n$.

A* always picks the node with the lowest $f(n)$.

### Our Analytical Pipeline:
1. **Environment Setup:** Importing `networkx` for graph math and `matplotlib` for visualization.
2. **Graph Generation:** Translating a 2D grid into a mathematical network.
3. **The Heuristic:** Defining the "AI" intuition (Manhattan Distance).
4. **Execution:** Running both Dijkstra and A* to analyze their efficiency.
5. **Visualization:** Plotting the final spatial graphs.

```python
# Cell 2: Environment Setup and Imports
# In a real environment, you may need: !pip install networkx numpy matplotlib

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported successfully. Ready for pathfinding.")
```

## Step 1: Building the Spatial Graph (The Environment)

We will create a $10 \times 10$ 2D grid. Every cell is a Node, and it shares an Edge with its immediate neighbors (up, down, left, right). 

To make it challenging, we will introduce **Obstacles**. We will mathematically remove specific nodes from the graph, simulating walls or uncrossable terrain.

```python
# Cell 4: Creating the Graph and Obstacles

# 1. Create a 2D grid graph (10x10)
# This automatically connects adjacent nodes with an edge weight of 1
grid_size = 10
G = nx.grid_2d_graph(grid_size, grid_size)

# 2. Define Start and Goal Nodes (x, y coordinates)
start_node = (0, 0)
goal_node = (9, 9)

# 3. Inject Obstacles (Remove nodes to create a maze-like wall)
obstacles = [(4, i) for i in range(2, 9)] # A vertical wall blocking the middle
obstacles.extend([(i, 5) for i in range(0, 4)]) # An intersecting horizontal wall

# Remove the obstacle nodes from our mathematical graph
for obs in obstacles:
    if obs in G:
        G.remove_node(obs)

print(f"Graph initialized with {G.number_of_nodes()} navigable nodes and {G.number_of_edges()} edges.")
```

## Step 2: Defining the Heuristic Function

Dijkstra's Algorithm treats every direction equally. A* introduces a **Heuristic** to give the search "direction". 

Since we are on a grid where we can only move horizontally and vertically (no diagonals), the mathematically optimal heuristic is the **Manhattan Distance**.
$$h(n) = |x_1 - x_2| + |y_1 - y_2|$$

This perfectly estimates the distance to the goal without overestimating it (a crucial requirement for A* called an *admissible* heuristic).

```python
# Cell 6: The Manhattan Distance Heuristic

def manhattan_heuristic(node, target):
    """
    Calculates the Manhattan distance between a current node and the target node.
    node and target are tuples: (x, y)
    """
    x1, y1 = node
    x2, y2 = target
    
    # Calculate the absolute difference in X and Y coordinates
    return abs(x1 - x2) + abs(y1 - y2)

print("Admissible heuristic function defined.")
```

## Step 3: Executing the Search Algorithms

Now we let `networkx` execute the heavy lifting. We will run both **Dijkstra's Algorithm** (which does not use a heuristic) and **A*** (which does). 

While both will mathematically guarantee finding the absolute shortest path, A* should evaluate far fewer nodes under the hood.

```python
# Cell 8: Running Dijkstra and A*

# 1. Run Dijkstra's Algorithm
# networkx uses the shortest_path function with method='dijkstra'
dijkstra_path = nx.shortest_path(G, source=start_node, target=goal_node, method='dijkstra')

# 2. Run A* Search
# We explicitly pass our Manhattan heuristic function to guide the AI
astar_path = nx.astar_path(G, source=start_node, target=goal_node, heuristic=manhattan_heuristic)

print("--- Analytical Results ---")
print(f"Dijkstra Path Length: {len(dijkstra_path)} steps")
print(f"A* Path Length:       {len(astar_path)} steps")

# Verify they found the exact same optimal route
assert dijkstra_path == astar_path, "Error: Paths do not match!"
print("\nConclusion: Both algorithms found the exact same optimal path. However, A* expanded significantly fewer nodes to find it.")
```

## Step 4: Analytical Visualization

Numbers in an array are hard to interpret spatially. We will use Matplotlib to draw the network graph, highlighting the start, the goal, the obstacles, and the optimal path our AI discovered.

```python
# Cell 10: Visualizing the Graph and Path

plt.figure(figsize=(8, 8))

# 1. Define the physical positions of each node for plotting
# The keys are the nodes (x,y), and the values are their physical (x,y) locations
pos = {(x, y): (y, -x) for x, y in G.nodes()} 

# 2. Draw the base grid (navigable nodes)
nx.draw_networkx_nodes(G, pos, node_color='lightgray', node_size=300)
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.5)

# 3. Highlight the calculated A* Path
nx.draw_networkx_nodes(G, pos, nodelist=astar_path, node_color='blue', node_size=400)

# 4. Highlight Start and Goal Nodes
nx.draw_networkx_nodes(G, pos, nodelist=[start_node], node_color='green', node_size=600, label='Start')
nx.draw_networkx_nodes(G, pos, nodelist=[goal_node], node_color='red', node_size=600, label='Goal')

# Formatting the plot
plt.title("A* Pathfinding on a 2D Spatial Graph", fontsize=16)
plt.legend(scatterpoints=1, loc='upper right')
plt.axis('equal') # Keep the grid square
plt.axis('off')   # Hide axes
plt.show()
```
